```python
#Correlacion

plt.figure(figsize=(13, 10))
sns.histplot(df_sal.salary_in_usd); # no distribuye normal, esta acumulado hacia los valores chicosm, ideal utilizar el metodo de correlacion de Kendall
```

```python
plt.figure(figsize=(13, 10))
for j, i in enumerate(['pearson','kendall','spearman']):
    plt.subplot(1,3,j+1)
    correlacion = df_sal.corr(method=i)
    sns.heatmap(correlacion)
    plt.title(i)
    plt.yticks(fontsize = 14)
    plt.xticks(fontsize = 14)
```

el mejor grafico pareciera ser el de kendall pero no muestra correlaciones significantes

la mayor correlacion (aparte de la de salarios y salarios en dolares que no tiene relevancia) es la del año y salario en dolares con una relacion del 21.5%

```python
#diferentes correlaciones

correlacion = df_sal.corr(method='kendall') 
v_correlacion = correlacion[correlacion !=1]
v_correlacion.unstack().drop_duplicates().reset_index().dropna().sort_values(0, ascending=False)
```

```python
plt.figure(figsize=(14, 14))

g = sns.pairplot(df_sal[['salary_in_usd', 'work_year', 'remote_ratio']], diag_kind='kde');
g.fig.suptitle("Grafico de puntos en variables numericas", fontsize=20, y=1.02)
```

# Categorica vs Numerica


```python
#salario vs nivel de experiencia

plt.figure(figsize=(15,10))
sns.violinplot(x = 'experience_level', y='salary_in_usd', data=df_sal,order=["EN", "MI", "SE", "EX"])
```

Acá si vemos que a mayor nivel de experiencia aumenta la cantidad de empleados con un mayor salario

```python
# salario vs tamaño de la compañia
plt.figure(figsize=(15,10))
sns.violinplot(x = 'company_size', y='salary_in_usd', data=df_sal,order=["S", "M", "L"])
```

Tambien hay relacion positiva entre tamaño de empresa y salarios altos. En cuanto a mayor parte de las personas, se puede decir que no hay demasiada diferencia entre salarios dependiendo del tamaño de la compañía

# Categórica vs Categórica


```python
# Tamaño de empresas vs nivel de experiencia

cross = pd.crosstab(df_sal['company_size'] , df_sal['experience_level'])
cross
cross.plot(kind="bar", stacked=True, rot=0, figsize=(15,12));
```

Se puede notar que las empresas chicas tienen pocos expertos y que las compañias medianas emplean muchos seniors
